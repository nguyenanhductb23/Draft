```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def gaussian(x, mean=0, sigma=1):
    return 1 / (np.sqrt(2*math.pi) * sigma) * np.exp(-0.5*((x - mean) / sigma)**2)
```

```python
x = np.arange(-5,5,0.01)
y = gaussian(x)
plt.plot(x, y)
plt.show()
```
