```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
T = 0.1
t = np.arange(0,1,0.05)

X = np.sin(2*math.pi*3*t)
plt.plot(t, X)
```

```python
t_new = np.arange(0.011,1.011,0.01)
X_new = np.zeros(t_new.size)

for n in range(len(X)):
    X_new = X_new + X[n] * np.sin(math.pi/T*(t_new-n*T)) / (math.pi/T*(t_new-n*T))
    
    
plt.plot(t_new, X_new)
```
